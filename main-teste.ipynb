```python
import pandas as pd
```

```python
df = pd.read_excel('Entradas - Cerne.xlsx')
df = df.drop(columns=['Eixo do CERNE 2', 'Eixo do CERNE 3'])

# Variávveis do Eixo de Cerne 1
CERNE_1 = 'Eixo do CERNE 1'
sensibilizacao = 'Sensibilização'
prospeccao = 'Prospecção'
Potenciais_Empreendedores = 'Sensibilização / Qualificação de Potenciais Empreendedores'
desenvolvimento_empreendedor = 'Desenvolvimento do Empreendedor'

# Tratamento dos dados do Eixo de Cerne 1
def tratamento_cerne_1():
    df[CERNE_1].dropna(inplace=True)
    df[CERNE_1] = df[CERNE_1].str.strip()
    df['Tipo de Entrada'].str.strip()
    df_tratado = df.dropna(subset=[CERNE_1]).reset_index(drop=True)
    df_tratado[CERNE_1] = df_tratado[CERNE_1].replace({
        'Sensibilização / \nQualificação de Potenciais Empreendedores': 'Sensibilização / Qualificação de Potenciais Empreendedores',
    })
    return df_tratado

df = tratamento_cerne_1()

#filtro no dataframe 
def filtro(valor_filtrado, coluna=CERNE_1):
    df_filtro = df[coluna] == valor_filtrado
    return df_filtro

# Recebe um filtro e uma coluna
def filter_Count(valor_filtrado, coluna=CERNE_1):
    total = df.loc[filtro(valor_filtrado), coluna]
    return total

def filtro_multiplo(filtro_1, filtro_2, coluna_2, coluna=CERNE_1):
    df_filtro = (filtro(filtro_1, coluna) & filtro(filtro_2, coluna_2))
    return df_filtro

def filtro_mult(filtro_1, filtro_2, coluna_2, coluna_3, coluna=CERNE_1):
    f1 = df[coluna] == filtro_1
    f2 = df[coluna_2] == filtro_2

    df_filtro = df.loc[f1 & f2, coluna_3]
    return df_filtro

print(df['Tipo de Entrada'].unique())
```
